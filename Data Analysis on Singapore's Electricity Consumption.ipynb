```python
#Removing na variables
plot = df2.dropna()
plot
```

```python
#Renaming of single column
plot.columns.values[0] = "Details"
plot
```

```python
#Selection of necessary columns for plot drawing, average of annual
plotannual = plot[["Details", "Annual"]]
plotannual
```

```python
#Annual data goes by chronological order from 2005, renaming of columns
plotannual.columns = ["Details", 2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]
plotannual
```

```python
#Selection of rows that reflect usage of different household types
overall = plotannual.loc[plotannual["Details"] == "Overall"]
overall
```

```python
#Removal of column
plotstart = overall.drop(["Details"], axis =1)
plotstart
```

```python
plotstart.reset_index(drop = True, inplace = True)
#row 0 refers to Overall
#row 1 is public housing 
#row 2 is 1-room/2-room
#row 3 is 3-room
#row 4 is 4-room
#row 5 is 5-room and Executive
#row 6 is Private Housing 
#row 7 is Landed Properties
#row 8 is Private Apartments and Condominiums
```

```python
#changing rows to columns and vice versa, renaming of columns 
table = plotstart.transpose()
table.columns = ["Average", "Public Housing", "1-room/2-room", "3-room", "4-room", "5-room and Executive", "Private Housing", "Landed Properties", "Private Apartments and Condominiums"]
table.reset_index(inplace = True)
table = table.rename({"index":"Year"}, axis = 1)
table
```

```python
#Graph plotting of Average electricity usage of Public and Private housing over the years 
housingtype = table.plot(x="Year", y = ["Average","Public Housing","Private Housing"])
housingtype.set_ylabel("Average kWh")
```

```python
#Graph plotting of Average electricity usage of different housing types over the years 
housingsize = table.plot(x="Year", y= ["1-room/2-room", "3-room", "4-room", "5-room and Executive", "Landed Properties", "Private Apartments and Condominiums"])
housingsize.set_ylabel("Average kWh")
plt.legend(bbox_to_anchor=(1.0, 1.0))
```
